```python
ventas_a = ventas_a.reduceByKey(lambda x,y: x+y)
```

```python
result = ventas_a.join(formato)
# id_video, (ventas totales, titulo)
```

```python
result_a = result.map(lambda x: (x[0],x[1][1],x[1][0]))
# id_video,  titulo, ventas totales
```
