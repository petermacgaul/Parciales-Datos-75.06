```python
import pandas as pd
data = pd.DataFrame([
    [18,11,2019,160,13],
    [18,11,2019,160,13],
    [18,11,2019,160,13],
    [18,11,2019,160,13],
    [18,11,2019,160,13],
    [18,11,2019,160,16],
    [18,10,2019,1,13],
    [18,10,2019,1,13],
    [18,10,2019,1,13],
    [18,10,2019,1,13],
    [18,10,2019,1,13],
    [18,10,2019,1,12],
    [18,3,2019,2,13],
    [18,10,2019,3,13],
    [18,10,2019,3,130]               
                    ],
columns=["dia", "mes", "ao", "prod id", "importe"])

data2 = pd.DataFrame([
    [1,"asd","Men",160],         
    [2,"asd","Women",160],         
    [3,"asd","Men",160],         
    [160,"asd","Men",160],         
                    ],
columns=["prod id", "title", "cat", "descript"])
```

```python
data_men = data2[data2["cat"] == "Men"]
```

```python
merged = pd.merge(data_men,data,on="prod id",how="inner")
```

```python
merged["avg"] = merged.groupby("mes").importe.transform(lambda x: x.mean())
merged
```

```python
result = merged[merged["avg"] < merged["importe"]]
```
