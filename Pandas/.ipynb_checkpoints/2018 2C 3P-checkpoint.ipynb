```python
import pandas as pd
data = pd.DataFrame([[18,"nico","ARG", "2019-11-18", 1000,160,13],
                     [17,"nico","ARG", "2019-11-18", 1000,149,13],
                     [16,"nico","ARG", "2019-11-18", 100,1,13],
                     [15,"nico","ARG", "2019-11-18", 100,3,13],
                     [14,"nico","ARG", "2019-11-18", 900,30,13],
                     [13,"nico","ARG", "2019-11-18", 801,200,13],
                     [12,"nico","RT", "2019-11-18", 1200,303,13],
                     [11,"nico","RT", "2019-11-18", 100,30,13]],
                     columns=["id_usuario", "nickname", "pais", "fecha_alta", "seguidores", "seguidos", "cant total public"])
```

```python
data
```

```python
data["max"] = data.groupby("pais").seguidores.transform(lambda x: x.max())
data
```

```python
data = data[(data["seguidores"]) > (data["max"]*0.8)]
```

```python
data = data[ data["seguidores"]*0.15 > data["seguidos"]]
```
